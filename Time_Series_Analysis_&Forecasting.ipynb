```python
df_google = df_google.reset_index()
df_google.info()
```

```python
pd.to_datetime(df_google['Date'])
```

```python
df_google = df_google.set_index('Date',drop = 'True')
```

```python
df_google.head()
```

## Time Resampling

```python
df_google.head()
```

```python
df_google.resample(rule = 'A').max()
```

There are many types of sampling rule check the directory

```python
df_google.resample(rule = 'A').max()['Open'].plot()
```

```python
df_google.resample(rule = 'A').min()
```

it give minimum values in the top.

```python
df_google["Open"].resample(rule = 'A').mean().plot(kind = 'bar')
```

```python
df_google['Open'].resample(rule = 'M').max().plot(kind = 'bar',figsize = (15,6))
```

## Rolling can be used for smoothing the curve
it is kind of aggregate column

```python
df_google['High'].rolling(11).max().head(20)
```

```python
df_google.head()
```

```python
df_google['open:30 days rolling'] = df_google['Open'].rolling(30).mean()
```

```python
df_google.head(31)
```

```python
df_google[['Open','open:30 days rolling']].plot(figsize=(12,5))
```

# orange curve is with 30 days rolling mean which give smoothen curve
